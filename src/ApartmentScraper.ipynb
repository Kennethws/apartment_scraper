```python
import requests
from bs4 import BeautifulSoup
import datetime
import sched
import webbrowser
import time
import tkinter as tk
from tkinter import *
from tkinter import ttk, messagebox, Toplevel, Label, Text, END

class ApartmentScraper:
    def __init__(self):
        self.url = "https://marketcentral.com/floor-plans"
        self.is_running = False
        self.scheduler = sched.scheduler(time.time, time.sleep)

    @staticmethod
    def open_link(event, text):
        start = text.index(tk.CURRENT)
        end = text.index(f"{start} wordend")
        url = text.get(start, end)
        webbrowser.open(url)

    def create_popup(self, message, suitable_apartments):
        def open_link(event):
            self.open_link(event, text)

        root = tk.Tk()
        root.withdraw()

        popup = Toplevel(root)
        popup.title("Suitable Apartments Found")
        popup.attributes("-topmost", True)

        text = Text(popup, wrap='word', padx=20, pady=20)
        text.insert(END, message)
        text.tag_configure("link", foreground="blue", underline=True)
        text.tag_bind("link", "<Button-1>", open_link)

        for apt in suitable_apartments:
            start_index = f"1.0 + {message.index(apt[7])} chars"
            end_index = f"{start_index} + {len(apt[7])} chars"
            text.tag_add("link", start_index, end_index)

        text.config(state='disabled')
        text.pack()

        ok_button = tk.Button(popup, text="OK", command=root.destroy, width=10)
        ok_button.pack(pady=10)

        root.mainloop()


    def find_studio_apartments(self):
        if not self.is_running:
            return

        response = requests.get(self.url)
        soup = BeautifulSoup(response.content, 'html.parser')

        listings_table = soup.find('table', {'id': 'availability'})
        rows = listings_table.find_all('tr', class_='data')

        suitable_apartments = []
        for row in rows:
            cells = row.find_all('td')
            unit = cells[0].text.strip()
            property_name = cells[1].text.strip()
            floor_plan = cells[2].text.strip()
            rent = cells[4].text.strip()
            bed_bath = cells[5].text.strip()
            sq_ft = cells[6].text.strip()
            available_date = datetime.datetime.strptime(row['data-available'], "%m/%d/%y")

            if property_name == "Watermark" and 'S' in floor_plan and available_date > datetime.datetime(2023, 7, 1):
                link = cells[-1].find('a', href=True, alt="Apply")['href']
                suitable_apartments.append((unit, property_name, floor_plan, rent, bed_bath, sq_ft, available_date.strftime("%m/%d/%y"), link))

        if suitable_apartments:
            message = "Suitable apartments found:\n\n"
            for apt in suitable_apartments:
                message += f"Unit: {apt[0]}\nProperty: {apt[1]}\nFloor Plan: {apt[2]}\nRent: {apt[3]}\nBed/Bath: {apt[4]}\nSq Ft: {apt[5]}\nAvailable: {apt[6]}\nLink: {apt[7]}\n\n"
            print(self.url)
            print()
            print(message)
            self.create_popup(message, suitable_apartments)
        else:
            print("No suitable apartments found.")

        # Schedule the function to run again after the chosen interval
        self.scheduler.enter(self.refresh_interval, 1, self.find_studio_apartments, ())

    def start(self, refresh_interval):
        self.refresh_interval = refresh_interval
        self.is_running = True

        # Run the function for the first time immediately
        self.find_studio_apartments()

        # Schedule the function to run periodically
        self.scheduler.enter(self.refresh_interval, 1, self.find_studio_apartments, ())
        self.scheduler.run()


    def stop(self):
        self.is_running = False


def start_scraper():
    refresh_interval = int(refresh_frequency.get())
    scraper.start(refresh_interval)


def stop_scraper():
    scraper.stop()


scraper = ApartmentScraper()

# Create the user interface
root = Tk()
root.title("Apartment Scraper")

mainframe = ttk.Frame(root, padding="12 12 12 12")
mainframe.grid(column=0, row=0, sticky=(N, W, E, S))

refresh_frequency = StringVar()
ttk.Label(mainframe, text="Refresh Frequency (seconds)").grid(column=1, row=1, sticky=W)
ttk.Entry(mainframe, width=7, textvariable=refresh_frequency).grid(column=2, row=1, sticky=(W, E))

start_button = ttk.Button(mainframe, text="Start", command=start_scraper)
start_button.grid(column=1, row=3, sticky=W)
stop_button = ttk.Button(mainframe, text="Stop", command=stop_scraper)
stop_button.grid(column=2, row=3, sticky=W)

root.mainloop()
```
